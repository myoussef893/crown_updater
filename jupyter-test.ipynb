```python
import pandas as pd 

date_list = pd.date_range('2025-01-01', periods=365.25, freq='D')
date_df = pd.DataFrame(date_list, columns=['date'])
hotel_df = pd.read_csv('hotel_index.csv')

merge_df = hotel_df.merge(date_df, how='cross')
merge_df['date'] = pd.to_datetime(merge_df['date'])
merge_df.to_sql('merge_df', con='sqlite:///hotel_data.db', if_exists='replace', index=False)

s_df = pd.read_sql('select * from staying', con='sqlite:///database.db')
s_df.Property.unique()

merge_df
```

```python
s_df.drop_duplicates(keep='last', inplace=True)
```

```python
s_df = pd.read_sql('select * from staying', con='sqlite:///database.db')
prop = 'Holland Road Stays'

merged = pd.read_sql(f'select * from merge_df where date < "2025-06-01" and "Property" = "{prop}"', con='sqlite:///hotel_data.db')
s_df.drop_duplicates(keep='last', inplace=True)
s_df['Check In'] = pd.to_datetime(s_df['Check In'])
s_df['Check Out'] = pd.to_datetime(s_df['Check Out'])
merged['date'] = pd.to_datetime(merged['date'])
merged['stay'] = 0

for i, row in merged.iterrows():
    # كام إقامة بتحصل في اليوم ده؟
    stay_count = ((s_df['Check In'] <= row['date']) & (row['date'] < s_df['Check Out']) & (row['Property'] == s_df['Property'])).sum()
    merged.at[i, 'stay'] = stay_count

merged['Occupancy'] = (merged.stay / merged['Total Rooms'])*100
```

```python
grand_table = merged
prop = 'Holland Road Stays'
staying_date = '2025-06-01'
staying_date_df = pd.read_sql(f'select * from staying where "Check Out" < "{staying_date}" and "Property" = "{prop}"', con = 'sqlite:///database.db') 
staying_date_df['Check In'] = pd.to_datetime(staying_date_df['Check In'])
staying_date_df['Check Out'] = pd.to_datetime(staying_date_df['Check Out'])

grand_table['Resv Count'] = 0

for i, row in grand_table.iterrows(): 

    stay_count = ((staying_date_df['Check In'] <= row['date']) & 
                (row['date'] < staying_date_df['Check Out']) & 
                (staying_date_df['Property'] == row['Property'])).sum()
    grand_table.at[i, 'Resv Count'] = stay_count



grand_table[grand_table['Resv Count']>2]
```
